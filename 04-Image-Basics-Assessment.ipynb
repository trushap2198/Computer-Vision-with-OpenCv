#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
fix=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
vert=np.array([[400,300],[100,700],[690,700]])
vert=vert.reshape(-1,1,2)
cv2.fillPoly(fix,[vert],(0,0,255))
plt.imshow(fix)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
imgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imgb)
```

```python
imgb=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
def draw_cir(event,x,y,flags,param):
    if event== cv2.EVENT_FLAG_RBUTTON:
        cv2.circle(imgb,(x,y),100,(0,0,255),10)
        
cv2.namedWindow(winname="draw")

cv2.setMouseCallback("draw",draw_cir)
while True:
    cv2.imshow("draw",imgb)
    if cv2.waitKey(20) & 0XFF==27:
        break;
cv2.destroyAllWindows()
plt.imshow(imgb)
```
